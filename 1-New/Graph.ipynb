```python
import time
import random
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker


# ─────────────────────────────────────────────
# 1. Tower of Hanoi  –  O(2^n)
# ─────────────────────────────────────────────
def tower_of_hanoi(n, src='A', dst='C', aux='B'):
    if n == 1:
        return
    tower_of_hanoi(n - 1, src, aux, dst)
    tower_of_hanoi(n - 1, aux, dst, src)


# ─────────────────────────────────────────────
# 2. First Duplicate  –  O(n^2)  (naïve nested-loop version)
# ─────────────────────────────────────────────
def first_duplicate(arr):
    for i in range(len(arr)):
        for j in range(i + 1, len(arr)):
            if arr[i] == arr[j]:
                return arr[i]
    return -1


# ─────────────────────────────────────────────
# 3. 0-1 Knapsack  –  O(n * W)
# ─────────────────────────────────────────────
def knapsack_01(weights, values, capacity):
    n = len(weights)
    dp = [[0] * (capacity + 1) for _ in range(n + 1)]
    for i in range(1, n + 1):
        for w in range(capacity + 1):
            dp[i][w] = dp[i - 1][w]
            if weights[i - 1] <= w:
                dp[i][w] = max(dp[i][w], dp[i - 1][w - weights[i - 1]] + values[i - 1])
    return dp[n][capacity]


# ─────────────────────────────────────────────
# 4. LCS (Longest Common Subsequence)  –  O(m * n)
# ─────────────────────────────────────────────
def lcs(s1, s2):
    m, n = len(s1), len(s2)
    dp = [[0] * (n + 1) for _ in range(m + 1)]
    for i in range(1, m + 1):
        for j in range(1, n + 1):
            if s1[i - 1] == s2[j - 1]:
                dp[i][j] = dp[i - 1][j - 1] + 1
            else:
                dp[i][j] = max(dp[i - 1][j], dp[i][j - 1])
    return dp[m][n]


# ─────────────────────────────────────────────
# 5. Sum of Subsets  –  O(2^n)  (recursive backtracking)
# ─────────────────────────────────────────────
def sum_of_subsets(arr, target, idx=0, current=0):
    if current == target:
        return True
    if idx == len(arr) or current > target:
        return False
    return (sum_of_subsets(arr, target, idx + 1, current + arr[idx]) or
            sum_of_subsets(arr, target, idx + 1, current))


# ─────────────────────────────────────────────
# Benchmarking helpers
# ─────────────────────────────────────────────
REPEATS = 5          # average over N runs to reduce noise


def measure(fn, *args):
    """Return average elapsed time in microseconds."""
    total = 0
    for _ in range(REPEATS):
        start = time.perf_counter()
        fn(*args)
        total += time.perf_counter() - start
    return (total / REPEATS) * 1_000_000   # → µs


def benchmark_hanoi(sizes):
    return [measure(tower_of_hanoi, n) for n in sizes]


def benchmark_first_duplicate(sizes):
    times = []
    for n in sizes:
        arr = list(range(n)) + [random.randint(0, n - 1)]
        random.shuffle(arr)
        times.append(measure(first_duplicate, arr))
    return times


def benchmark_knapsack(sizes):
    times = []
    for n in sizes:
        capacity = n * 10
        weights = [random.randint(1, 20) for _ in range(n)]
        values  = [random.randint(1, 50) for _ in range(n)]
        times.append(measure(knapsack_01, weights, values, capacity))
    return times


def benchmark_lcs(sizes):
    times = []
    chars = 'abcdefghijklmnopqrstuvwxyz'
    for n in sizes:
        s1 = ''.join(random.choices(chars, k=n))
        s2 = ''.join(random.choices(chars, k=n))
        times.append(measure(lcs, s1, s2))
    return times


def benchmark_sum_of_subsets(sizes):
    times = []
    for n in sizes:
        arr    = [random.randint(1, 20) for _ in range(n)]
        target = random.randint(sum(arr) // 4, sum(arr) // 2)
        times.append(measure(sum_of_subsets, arr, target))
    return times


# ─────────────────────────────────────────────
# Input-size ranges  (tune as needed)
# ─────────────────────────────────────────────
configs = {
    "Tower of Hanoi – O(2ⁿ)":               (range(1, 26),   benchmark_hanoi),
    "First Duplicate – O(n²)":               (range(50, 1001, 50), benchmark_first_duplicate),
    "0-1 Knapsack – O(n·W)":                 (range(10, 201, 10),  benchmark_knapsack),
    "LCS – O(m·n)":                          (range(50, 1001, 50), benchmark_lcs),
    "Sum of Subsets – O(2ⁿ)":               (range(1, 26),   benchmark_sum_of_subsets),
}


# ─────────────────────────────────────────────
# Plot – one subplot per algorithm
# ─────────────────────────────────────────────
fig, axes = plt.subplots(2, 3, figsize=(16, 10))
fig.suptitle("Algorithm Performance Benchmarks", fontsize=16, fontweight='bold', y=1.01)
axes = axes.flatten()

for ax, (title, (size_range, bench_fn)) in zip(axes, configs.items()):
    sizes = list(size_range)
    print(f"Benchmarking: {title} …")
    times = bench_fn(sizes)

    ax.plot(sizes, times, marker='o', markersize=4, linewidth=1.8, color='steelblue')
    ax.set_title(title, fontsize=11, pad=8)
    ax.set_xlabel("Input size (n)", fontsize=9)
    ax.set_ylabel("Time taken (µs)", fontsize=9)
    ax.yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, _: f"{x:,.0f}"))
    ax.grid(True, linestyle='--', linewidth=0.5, alpha=0.7)
    ax.set_facecolor('#f9f9f9')

# Hide the unused 6th subplot
axes[-1].set_visible(False)

plt.tight_layout()
plt.savefig("algorithm_benchmarks.png", dpi=150, bbox_inches='tight')
print("\nSaved → algorithm_benchmarks.png")
plt.show()
```
